## Using the IO-Model-Builder
`iomb` is an open source Python library for creating environmentally extended input-output models (EEIO models) from CSV files. This notebook shows the functionalities of `iomb` with a simple example.

### Calculating a coefficients matrix
The calculation and conversion functions of `iomb` are based on a direct requirements coefficients matrix. `iomb` contains functions to obtain such a matrix from supply and use tables as described in the 
[Concepts and Methods of the U.S. Input-Output Accounts, Chapter 12][1]. 

[1]:http://www.bea.gov/papers/pdf/IOmanual_092906.pdf "Karen J. Horowitz, Mark A. Planting: Concepts and Methods of the U.S. Input-Output Accounts. 2006"


```python
import iomb

# read an input output model from a supply and use table)
io_model = iomb.make_io_model('supply_table.csv', 'use_table.csv')
```

A use table is typically a commodity\*industry table with additional rows for added values and columns for final demands. The use table of our example looks like this:

```python
io_model.use_table
```
